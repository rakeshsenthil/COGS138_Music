## Background & Prior Work

Music is a universal part of the human experience and appears in every human society observed.(Gottlieb) It is known to be able to provoke strong emotions and influence moods in people. Because of its ability to influence mood and perception, music has been used as part of therapy for managing both physical pain and mental health.(Guo, Lu) Music therapy in particular is a non-invasive and cost-effective therapy that has been shown to reduce pain and anxiety levels in several clinical populations, including burn patients.(Ettenberger)

Since music has the potential to evoke emotion, it has been used as a tool to investigate emotion and its neural correlates. One fMRI study that tried to induce happy and sad states using classical music found an increased BOLD signal in the ventral and dorsal striatum, anterior cingulate, parahippocampal gyrus, and auditory association areas when listening to happy music and in the hippocampus/amygdala and auditory association areas when listening to sad music (Mitterschiffthaler) Another fMRI study involving piano music found BOLD filtered tempo and velocity (loudness) curves for expressive performances were highly correlated and tempo fluctuation in expressive performances predicted emotional arousal ratings for the participants.(Chapin)

Specific features of music and their neural and emotional responses have also been studied. An EEG study found that musical stimuli retaining basic musical features, such as rhythm and melody, elicited significantly higher behavioral ratings and neural correlation from their participants than did phase-scrambled controls and frequency-domain measures of correlation (coherence) peaked at frequencies related to the musical beat (Kaneshiro)  Another EEG study that investigated the effect of tempo in the brain found that tempo had an effect on the direction and strength of motor-auditory interactions and that topological patterns of brain activity (this study focused on the imaginary part of coherence (iCOH) in the frequency range 1.5–18 Hz (δ, θ, α and lower β)) were similar for music and noise but that there were  statistically significant differences in strength and direction of iCOH identified (Nicolau)

# Results: MFCC

MFCC is short for Mel-frequency cepstral coefficients. From wikipedia “the frequency bands are equally spaced on the mel scale, which approximates the human auditory system's response more closely than the linearly-spaced frequency bands used in the normal spectrum. This frequency warping can allow for better representation of sound, for example, in audio compression.” By having some sort of librosa auditory feature, we wanted to see if theres any way we can predict these coefficients given an EEG signal. And Based on our results of the alpha/beta bands we found that there really isn’t a correlation. 

```python
from scipy.optimize import curve_fit

def linear_func(x, a, b):
    return np.multiply(x,a) + b

def fit_r2(ydata, yfit):
    residuals = ydata - yfit # difference between data and fit
    ss_res = np.sum(residuals**2) # sum of the squares (ss) of the residuals
    ss_tot = np.sum((ydata - np.mean(ydata))**2) # total sum of squares
    r2 = 1 - (ss_res / ss_tot) # r-squared    
    return r2

def quadratic_func(x,a,b,c):
    return np.multiply(np.power(x,2),a) + np.multiply(x,b) + c



# Because it can be non-intuitive, the below code will plot
# your data and the fits, so you can test your code out
#
```
